## BÀI 1: Một lớp học bắt đầu từ 8h và kết thúc lúc 10h. Giáo viên điểm danh vào thời điểm ngẫu nhiên trong khoảng thời gian  t=30  phút đầu. Một học sinh đến lớp vào thời điểm ngẫu nhiên trong giờ học (và học cho đến khi hết giờ).
### a) Tính xác suất học sinh được điểm danh.*
### b) Khảo sát xác suất học sinh được điểm danh theo thời gian đầu  t  phút  (0≤t≤120).*

Gọi $A$ là biến cố học sinh được điểm danh, $x$ là số phút tại thời điểm học sinh đến lớp trong lúc tiết học diễn ra: $ 0 \leqslant x \leqslant 120$, $y$ là số phút tại thời điểm tiết học diễn ra: $ 0 \leqslant y \leqslant 120 $

Như vậy số phút học sinh đến trong lúc tiết học đang diễn ra được biểu diễn bằng một điểm có tọa độ $(x, y)$ nằm trong hình vuông có cạnh $120$ (đơn vị đo: Phút). Đó chính là miền $W$

Không gian mẫu: $\Omega = \{(x,y): 0 \leqslant x, y \leqslant 120\}$, $|\Omega| = S_{OABC} = 120^2 = 14400 $

Biểu diễn x, y lên hệ trục tọa độ Descartes. Chọn gốc tọa độ là lúc 0 phút (tiết học bắt đầu).

Để học sinh được điểm danh thì số phút $ x, y $ phải thỏa điều kiện:

$ 0 \leqslant x, y \leqslant 30  (*) \\ |y - x|\leq 30 \Leftrightarrow \begin{cases}y - x \leq 30 \\y - x \geq - 30 \end{cases} \Leftrightarrow \begin{cases}y \leq x + 30 \\y \geq x - 30 \end{cases}  (**) $

Từ $(*)$ và $(**)$, trường hợp thuận lợi cho biến cố $A$ là các cặp điểm $(x, y)$ nằm trong $OMNP$ được tô màu xanh lá (như hình vẽ).

Theo công thức xác suất hình học ta có:

$ P(A) = { S _{OMNP} \over S _{OABC}} = 2.{ S _{\bigtriangleup ONP} \over S _{\bigtriangleup OBC}} = 2.{ \frac{\frac{1}{2}.30^2}{\frac{1}{2}.120^2}} = {1 \over 8} = 0.125 $

```python
from fractions import Fraction as F
P_A = 2 * F(30**2, 120**2)
print(P_A, float(P_A))
```

**Mô phỏng bằng code**

```python
import random
def xac_suat_diem_danh(khong_gian_mau, thoi_gian_diem_danh):
    duoc_diem_danh = 0
    i = 0
    while i < khong_gian_mau:
        #Chọn ngẫu nhiên thời gian điểm danh và thời gian đi học (đơn vị phút)
        if thoi_gian_diem_danh == 0: #Throw error randrange(0,0,0)
            diem_danh_luc = 0
        else:
            diem_danh_luc = random.randrange(0, thoi_gian_diem_danh)
        di_hoc_luc = random.randrange(0, 120)
        if(di_hoc_luc <= diem_danh_luc):
            duoc_diem_danh += 1
        i += 1
    return duoc_diem_danh/khong_gian_mau
```

## BÀI 2. Bộ bài tiến lên 52 lá, xào “đều”, chia cho 4 người chơi. Khả năng có người được 13 lá khác số là hiếm cỡ nào?

Không gian mẫu $\Omega = \{\text{Tổ hợp chia } 13 \text{ lá bài từ bộ bài } 52 \},$ $ |\Omega| = C_{52}^{13} = 635013559600$

Gọi $A$ là biến cố "Có được $13$ lá khác số"

Được 1 quân bài trong 4 quân bài đồng cơ số $ \text{(Cơ, rô, chuồn, bích)} $, ta có $ C_{4}^{1} = 4$

Để có được 13 lá bài khác số, thì người chơi phải được $ C_4^1(2) * C_4^1(3) * C_4^1(4)....C_4^1(Át) \Leftrightarrow (C_4^1)^{13} = 67108864$

$ \Rightarrow |A| = (C_4^1)^{13} = 67108864 $

Dùng mô hình xác suất đơn giản, ta có:

$$P(A) = \frac{|A|}{|\Omega|} = \frac{67108864}{635013559600} = 0.00015068$$

```python
from fractions import Fraction as F
P_A = F(67108864, 635013559600)
print(P_A, float(P_A))
```
